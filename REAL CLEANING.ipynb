```python
#observo que hay nulos,generos... voy a corregir algunos

sharksdf['Name'].fillna('Unknown', inplace=True)
sharksdf.loc[sharksdf['Name'].str.contains('male',na=False),'Name']='Unknown'
sharksdf.loc[sharksdf['Name'].str.contains('boy',na=False),'Name']='Unknown'
sharksdf.loc[sharksdf['Name'].str.contains('female',na=False),'Name']='Unknown'
sharksdf['Name']=sharksdf['Name'].str.replace('Anonymous','Unknown')
sharksdf['Name']=sharksdf['Name'].str.replace('Male','Unknown')
sharksdf['Name']=sharksdf['Name'].str.replace('fisherman','Unknown')
sharksdf['Name']=sharksdf['Name'].str.replace('Female','Unknown')
sharksdf['Name']=sharksdf['Name'].str.replace('woman','Unknown')
sharksdf['Name']=sharksdf['Name'].str.replace('sailor','Unknown')
sharksdf['Name']=sharksdf['Name'].str.replace('crewman','Unknown')
sharksdf['Name']=sharksdf['Name'].str.replace('a young Scotsman','Unknown')
sharksdf['Name']=sharksdf['Name'].str.replace('a native ','')
sharksdf['Name'] = sharksdf['Name'].replace('\s+', ' ', regex=True)
```

```python
#VAMOS A POR TYPE
set(sharksdf['Type']) # Buena pinta. Poco que limpiar.
```

```python
sharksdf['Type'] = sharksdf['Type'].str.strip()
sharksdf['Type'] = sharksdf['Type'].str.replace('Boating','Boat') #Boating por Boat
sharksdf['Type'] = sharksdf['Type'].str.replace('Invalid','Unknown') #Invalid por Unknown
sharksdf['Type'] = sharksdf['Type'].str.replace('Boatomg','Boat') #Boatomg por Boat
sharksdf['Type'].fillna('Unknown', inplace=True) #nulos por Unknown
set(sharksdf["Type"]) #así quedan los Type
```

```python
#COLUMNA SEX

set(sharksdf['Sex']) #Aquí solo quiero que aparezca F, M y UNKNOWN.Tal que así
```

```python
sharksdf['Sex'] = sharksdf['Sex'].str.replace('N', 'M') 
#La N y la M están muy pegadas en el teclado. Asumo que es un error de transcripción

sharksdf['Sex'] = sharksdf['Sex'].str.strip()
sharksdf['Sex'] = sharksdf['Sex'].fillna('UNKNOWN') # Relleno los nulos con UNKNOWN
sharksdf['Sex'] = sharksdf['Sex'].str.replace('lli', 'UNKNOWN') # lli ni idea de que quiere decir. Asi que UNKNOWN
sharksdf['Sex'] = sharksdf['Sex'].str.replace('.', 'UNKNOWN') # El punto no me dice nada
set(sharksdf['Sex']) # Así me quedaría
```

```python
# COLUMNA CASENUMBER

 #hay nulos, ND, y fechas seguidas de .a .b .....
```

```python
sharksdf['CaseNumber'] = sharksdf['CaseNumber'].str.replace('.c', '')
sharksdf['CaseNumber'] = sharksdf['CaseNumber'].str.replace('.R', '')
sharksdf['CaseNumber'] = sharksdf['CaseNumber'].str.replace('.b', '')
sharksdf['CaseNumber'] = sharksdf['CaseNumber'].str.replace('.a', '')
sharksdf['CaseNumber'] = sharksdf['CaseNumber'].str.replace('.f', '')
sharksdf['CaseNumber'] = sharksdf['CaseNumber'].str.replace('.e', '')
sharksdf['CaseNumber'] = sharksdf['CaseNumber'].str.replace('.h', '')
sharksdf['CaseNumber'] = sharksdf['CaseNumber'].str.replace('.R.b', '')
sharksdf['CaseNumber'] = sharksdf['CaseNumber'].str.replace('.R.a', '')
sharksdf['CaseNumber'] = sharksdf['CaseNumber'].str.replace('.d', '')
sharksdf['CaseNumber'] = sharksdf['CaseNumber'].str.replace('.k', '')
sharksdf['CaseNumber'] = sharksdf['CaseNumber'].fillna('UNKNOWN')
sharksdf['CaseNumber'] = sharksdf['CaseNumber'].str.replace('.i', '')
sharksdf['CaseNumber'] = sharksdf['CaseNumber'].str.replace('xx', 'UNKNOWN')
sharksdf['CaseNumber'] = sharksdf['CaseNumber'].str.replace(' g', '')

# He limpiado a piñón la mayoría de las filas que contenían errores
```
